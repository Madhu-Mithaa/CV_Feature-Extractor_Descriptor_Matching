# **Haris corner detection**

```python
# Importing the libraries
import cv2
import numpy as np

# Reading the image and converting the image to B/W
image = cv2.imread('/content/rgb_0023.png')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image = np.float32(gray_image)

# Applying the function
dst = cv2.cornerHarris(gray_image, blockSize=2, ksize=3, k=0.04)

# dilate to mark the corners
dst = cv2.dilate(dst, None)
image[dst > 0.01 * dst.max()] = [0, 255, 0]

```

```python
import matplotlib.pyplot as plt
plt.imshow(image),plt.show()
```

# **Histogram of Oriented Gradients - HOG**

```python
import cv2
import numpy as np 

#Reading the Image
imput_img = '/content/rgb_0023.png'
ori = cv2.imread(imput_img)
image = cv2.imread(imput_img)

#Converting to B/W
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

#Applying the function
dst = cv2.cornerHarris(gray,2,3,0.04)

# Dilate to mark the corners
dst = cv2.dilate(dst,None)
image[dst>0.01*dst.max()]=[0,0,255]
```

# **3.Feature Matching**

# **ORB**

```python
import cv2

#Read
img1 = cv2.imread('/content/rgb_0023.png', 0)
img2 = cv2.imread('/content/rgb_0184.png', 0)

#Applying the function
orb = cv2.ORB_create(nfeatures=500)

#Key Points
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

#Bf Matcher
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

matches = bf.match(des1, des2)
matches = sorted(matches, key=lambda x: x.distance)

match_img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:50], None)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1),plt.show()
plt.imshow(img2),plt.show()
plt.imshow(match_img),plt.show()
```

# **SIFT**

```python
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline

# read images
img1 = cv2.imread('/content/rgb_0023.png')  
img2 = cv2.imread('/content/rgb_0184.png') 

#Convert to B/W
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

#Applying the function
sift = cv2.xfeatures2d.SIFT_create()

#Key Points
keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)

#Feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1),plt.show()
plt.imshow(img2),plt.show()
plt.imshow(img3),plt.show()
```

# **SURF**

```python
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline

# read images
img1 = cv2.imread('/content/rgb_0023.png')  
img2 = cv2.imread('/content/rgb_0184.png') 

#Convert to B/W
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

#Applying the function
surf = cv2.xfeatures2d.SURF_create(400)

#Key Points
keypoints_1, descriptors_1 = surf.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = surf.detectAndCompute(img2,None)

#Feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1),plt.show()
plt.imshow(img2),plt.show()
plt.imshow(img3),plt.show()
```
